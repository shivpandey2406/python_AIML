```python
x = ['s', 'm', 's', 's', 'm', 'n', 's', 's', 'm', 'm', 's']
Char_count = {}
for char in x :
  if char in Char_count :
    Char_count[char] +=  1
  else:
    Char_count[char] = 1

print(Char_count)
```
