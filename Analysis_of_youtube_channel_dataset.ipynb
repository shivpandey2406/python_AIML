#Datacleaning (Replace '--' to NaN)

```python
data = data.replace('--',np.nan,regex=True)
data.head(20)
```

## Check null values in the dataset

```python
data.isnull().sum()
```

```python
per_missing = data.isnull().sum()*100/len(data)
per_missing
```

```python
sns.heatmap(data.isnull())
```

```python
data.dropna(axis=0,inplace = True)
sns.heatmap(data.isnull())
```

## Data cleaning Rank Column

```python
data.head()
```

```python
data.tail()
```

```python
# So now we remove the string values as well as comma
data.dtypes
## So we convert also object types into int types because ml used int types data for most cases
```

```python
data['Rank'] = data['Rank'].str[0:-2]
data.head()
```

```python
data['Rank'] = data['Rank'].str.replace(',','')
data.tail()
```

```python
data['Rank'] = data['Rank'].astype(int)
data.dtypes
```

## Data Cleaning onto Video uploads and Subscribers

```python
data.head()
data.dtypes  # Check the datatypes and convert object types into int types
data['Video Uploads'] = data['Video Uploads'].astype(int)
data['Subscribers'] = data['Subscribers'].astype(int)
data.dtypes
```

## Data cleaning grade column

```python
data.head()
data['Grade'].unique()
```

```python
data['Grade'] = data['Grade'].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1})
```

```python
data.head()
```

## Find the average views for each channel

```python
data.columns
```

```python
data['Avg_views'] = data['Video views']/data['Video Uploads']
data.head()
```

## Find out top 5 channels with maximum number of video uploads

```python
data.sort_values(by='Video Uploads' , ascending = False).head()
```
