## 8. Rename Month Names to English

```python
data['month_new'] = data['month'].replace(['Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro'],['January','February','March','April','May','June','July','August','September','October','November','December'])
```

# 9. Total Number of Fires Registered



```python
len(data['number'])
```

## 10. In Which Month Maximum Number of Forest Fires Were Reported?


```python
data.columns
```

```python
data1 = data.groupby('month_new')['number'].sum().sort_values(ascending= False)
data1
```

```python
plt.figure(figsize=(16,12))
sns.barplot(x=data1.index, y = data1.values, palette='rocket', color = 'orange')
```

11. In Which Year Maximum Number of Forest Fires Was Reported?

## 14. Display Number of Fires Were Reported in Amazonas (Year-Wise)



```python
data4 = data[data['state']=='Amazonas'].groupby('year')['number'].sum()  # But if we want to arrange in ascending order we need to add .sort_values(ascending= False)
data4
```

```python
sns.barplot(x=data4.index,y=data4.values,color = 'orange',palette = 'rocket')
plt.xticks(rotation= 75)
plt.show()
```

## 15. Display Number of Fires Were Reported in Amazonas (Day-Wise)

```python
data5 = data[data['state']=='Amazonas']
data5
```

```python
data5['date'] = pd.to_datetime(data5['date'])
data5.groupby(data5['date'].dt.dayofweek)['number'].sum()
```

```python
day = data5.groupby(data5['date'].dt.dayofweek)['number'].sum()
import calendar
day.index = [calendar.day_name[i] for i in range(0,7)]
day
```

```python
sns.barplot(x=day.index,y=day.values,color = 'orange',palette='rocket')
plt.xticks(rotation=80)
plt.show()
```

#16. Find Total Number of Fires Were Reported In 2015 And Visualize Data Based on Each ‘Month’

```python
Fire= data[data['year']==2015].groupby('month_new')['number'].sum()
Fire
```

```python
sns.barplot(x = Fire.index,y=Fire.values, color='orange',palette='rocket')
plt.xticks(rotation=75)
plt.show()
```

## 17. Find Average Number of Fires Were Reported from Highest to Lowest (State-Wise)



```python
data6 = data.groupby('state')['number'].mean().sort_values(ascending= False).reset_index()
data6
```

```python
sns.barplot(x="state",y="number",data=data6,color='orange',palette='rocket')
plt.xticks(rotation = 82)
plt.show()
```
